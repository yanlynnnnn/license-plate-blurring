```python
from tensorflow import keras
from Stats import Splitter
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from keras.callbacks import EarlyStopping
import pickle
```

```python
splitter = Splitter()
df = splitter.generate_features()
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns, index=df.index)
#pickle.dump(scaled_df, open('scaled_df.pkl', 'wb'))
#pickle.dump(df, open('split_df.pkl', 'wb'))
```

```python
# df = pickle.load(open('split_df.pkl', 'rb'))
df = pickle.load(open('scaled_df.pkl', 'rb'))
df = df.fillna(0)
df
```

```python
train_df = df[(df.index % 10 != 7) & (df.index % 10 != 8) & (df.index % 10 != 9)]
val_df = df[(df.index % 10 == 7) | (df.index % 10 == 8) | (df.index % 10 == 9)]
train_df = train_df.reset_index(drop=True)
val_df = val_df.reset_index(drop=True)
train_df
```

```python
train_data = train_df.to_numpy()
val_data = val_df.to_numpy()
train_X = train_data[:, :-1]
train_Y = train_data[:, -1]
val_X = val_data[:, :-1]
val_Y = val_data[:, -1]
```

```python
# One Hot Encoding for drivers 
label_encoder = LabelEncoder()
train_integer_encoded = label_encoder.fit_transform(train_Y)
val_integer_encoded = label_encoder.fit_transform(val_Y)
onehot_encoder = OneHotEncoder(sparse=False)
train_integer_encoded = train_integer_encoded.reshape(len(train_integer_encoded), 1)
val_integer_encoded = val_integer_encoded.reshape(len(val_integer_encoded), 1)
train_Y_encoded = onehot_encoder.fit_transform(train_integer_encoded)
val_Y_encoded  = onehot_encoder.fit_transform(val_integer_encoded)
```

```python
model = Sequential()
model.add(Dense(512, input_dim=64, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(137, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
early_stopping = EarlyStopping(monitor='loss', mode='min', patience=10)
```

```python
model.fit(train_X, train_Y_encoded, epochs=200, batch_size=20, validation_data=(val_X, val_Y_encoded), callbacks=[early_stopping])
```
