```python
from tensorflow import keras
from Stats import Splitter
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
splitter = Splitter()
df = splitter.generate_features()
```

```python
train_df = df[(df.index % 10 != 7) & (df.index % 10 != 8) & (df.index % 10 != 9)]
test_df = df[(df.index % 10 == 7) & (df.index % 10 == 8) & (df.index % 10 == 9)]
train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
train_df
```

```python
train_data = train_df.to_numpy()
train_X = train_data[:, :-1]
train_Y = train_data[:, -1]
```

```python
# One Hot Encoding for drivers 
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(train_Y)
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
train_Y_encoded = onehot_encoder.fit_transform(integer_encoded)
train_Y_encoded
```
